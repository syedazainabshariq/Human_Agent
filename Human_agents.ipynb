```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
def generate_response(prompt, max_length=100, temperature=0.7):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(input_ids, max_length=max_length, temperature=temperature, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the model and tokenizer
loaded_model = GPT2LMHeadModel.from_pretrained('my_gpt2_model')
loaded_tokenizer = GPT2Tokenizer.from_pretrained('my_gpt2_tokenizer')

# Define the response generation function
def generate_response(prompt, model, tokenizer, max_length=100, temperature=0.7):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(input_ids, max_length=max_length, temperature=temperature, pad_token_id=tokenizer.eos_token_id, do_sample=True)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response

# Generate response
prompt = "Explain how AI can revolutionize the future."
response = generate_response(prompt, model=loaded_model, tokenizer=loaded_tokenizer)
print(response)
```
